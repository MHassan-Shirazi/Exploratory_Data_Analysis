```python
df["Total_marks"] = df[['sql_marks', 'excel_marks' , 'python_marks' , 'power_bi_marks' , 'english_marks']].sum(axis = 1)
```

```python

df
```

```python
df['Percentage'] = df['Total_marks'] / 500 * 100
```

```python
df
```

```python
def grade(Percentage):

  if Percentage >= 90:
    return "A Grade"
  elif Percentage >= 80:
    return "B Grade"
  else:
    return "Fail"
```

```python
df['Grades'] = df['Percentage'].apply(grade)
```

```python
df
```

```python
Grades = ['A Grade' , 'B Grade' , 'Fail']
```

```python
grade_counts = df['Grades'].value_counts().reindex(Grades)
```

```python
grade_counts = grade_counts.sort_values(ascending=False)
```

```python
grade_counts
```

```python
plt.bar(grade_counts.index, grade_counts.values , color = ['Blue' , 'Red' , 'Orange'])
plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.title('Students Grades')
plt.xticks(rotation = 45)

plt.show()
```
